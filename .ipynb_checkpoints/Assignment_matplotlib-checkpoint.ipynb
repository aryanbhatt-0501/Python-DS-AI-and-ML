```python
import numpy as np;
import pandas as pd;
import matplotlib.pyplot as plt;
```

```python
df = pd.read_csv("product_sales.csv");
df.head()
```

```python
df.describe()
```

```python
# Sales trend of product A over the 12 months
df["Date"] = pd.to_datetime(df["Date"]);
df["Month"] = df["Date"].dt.to_period("M");

trend_A = df.groupby("Month")["Product A"].mean();
# print(trend_A);
plt.plot(trend_A.index.astype(str), trend_A.values, marker="o", linestyle="-");
plt.grid(True);
plt.xlabel("Months");
plt.ylabel("Sale");
plt.title("Sales trend of product A over the 12 months");
plt.show();
```

```python
# Plot the sales trend of all the products in a single graph
monthly_sales = df.groupby("Month")[["Product A", "Product B", "Product C"]].mean();
months = monthly_sales.index.astype(str);
plt.figure(figsize=(10,5))
plt.plot(months, monthly_sales["Product A"], marker="o", linestyle="solid", label="Product A")
plt.plot(months, monthly_sales["Product B"], marker="s", linestyle="dotted", label="Product B")
plt.plot(months, monthly_sales["Product C"], marker="*", linestyle="dashed", label="Product C")
plt.legend(loc="upper right", bbox_to_anchor=(1, .8))
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)

plt.show()
```

```python
# Use the total yearly sales of all products to create a bar chart
df.head()
```

```python
product_A = df["Product A"].sum();
product_B = df["Product B"].sum();
product_C = df["Product C"].sum();

products = ["Product A", "Product B", "Product C"];
values = [product_A, product_B, product_C];
colors = ["skyblue"] * len(products);
max_index = values.index(max(values));
colors[max_index] = "orange";
plt.style.use('seaborn-v0_8-white')
plt.bar(x=products, height=values, color=colors)
plt.title("Sales numbers over the years")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.tight_layout()
plt.show()
```

```python
fluctuations = np.random.randn(30);
plt.style.use('ggplot')
plt.figure(figsize=(8, 5))
plt.hist(fluctuations, bins=10, edgecolor="black")
plt.title("Simulated Daily Demand Fluctuations")
plt.xlabel("Fluctuation Amount")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```
