```python
import numpy as np;
import pandas as pd;
```

```python
df = pd.read_csv('light_spotify_dataset.csv');
```

```python
df.head()
```

```python
# Check for null values
missing_values = df.isnull().sum()
print('Missing values: ', missing_values[missing_values>0]);
```

```python
df
```

```python
df_cleaned = df.dropna(subset=['song'])
```

```python
df_cleaned
```

```python
df_cleaned.dtypes
```

```python
df_cleaned['Release Date'] = pd.to_datetime(df_cleaned['Release Date'], format="%Y");
df_cleaned.dtypes
```

```python
df_cleaned.head()
```

```python
# Duplicate values
duplicate_rows = df_cleaned.duplicated().sum();
print(duplicate_rows);
```

```python
df_cleaned.drop_duplicates(inplace=True)
```

```python
emotions = df_cleaned.groupby('emotion')['Popularity'].mean().sort_values(ascending=False);
print(emotions)
```

```python
byGenre = df_cleaned.groupby('Genre')[['Energy', 'Danceability']].mean().sort_values(by = ['Energy', 'Danceability'],ascending=[False, False]);
print(byGenre);
```

```python
# Top artist by songs count
max_count = df_cleaned.groupby('artist')['song'].count().sort_values(ascending=False);
print(max_count)
```

```python
# Number of songs released in an year by all the artists
number_of_songs = df_cleaned.groupby(['Release Date', 'artist'])['song'].count().sort_values(ascending=False);
print(number_of_songs);
```

```python
# Loudest artist every year
loudest_artist_year = df_cleaned.groupby(['Release Date','artist'])['Loudness'].max().sort_values(ascending=False);
print(loudest_artist_year);
# print(df['Loudness'].max());
```

```python
# Number of explicit songs by each artist
explicit_songs = df_cleaned[df_cleaned['Explicit'] == 'Yes'].groupby('artist').size().sort_values(ascending=False);
print(explicit_songs);
```

```python
# Popular songs of each artist
popular_songs = df_cleaned.groupby(['artist', 'song'])['Popularity'].mean().sort_values(ascending=False);
print(popular_songs);
```
