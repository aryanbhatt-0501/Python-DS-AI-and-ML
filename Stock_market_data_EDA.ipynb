### Task 3 - Basic analysis

```python
df['Price Change'] = df['Close'] - df['Open']
```

```python
df.head()
```

```python
percentChangeOfClosing = df['Close'].pct_change() * 100
```

```python
percentChangeOfClosing
```

```python
sns.lineplot(data=percentChangeOfClosing)
plt.xlabel('Days')
plt.ylabel('% change in closing price')
plt.show()
```

```python
print('Maximum high',df2.loc[df['High'].idxmax(), 'Date'])
print('Maximum low',df2.loc[df['Low'].idxmax(), 'Date'])
print('Maximum close',df2.loc[df['Close'].idxmax(), 'Date'])
print('Minimum high',df2.loc[df['High'].idxmin(), 'Date'])
print('Minimum low',df2.loc[df['Low'].idxmin(), 'Date'])
print('Minimum close',df2.loc[df['Close'].idxmin(), 'Date'])
```

### Task 4 - Basic Visualization 

```python
import matplotlib.dates as mdates
plt.figure(figsize=(20,6))
sns.lineplot(data=df2, x='Date', y='Close')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d-%b'))
plt.xticks(rotation=90)
plt.ylabel('Closing price')
plt.title('Price trend')
plt.show()
```
