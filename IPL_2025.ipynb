```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
deliveryPath = "IPL2025/deliveries.csv";
matchesPath = "IPL2025/matches.csv";
orange_capPath = "IPL2025/orange_cap.csv";
purple_capPath = "IPL2025/purple_cap.csv";
```

```python
delivery_dataset = pd.read_csv(deliveryPath);
matches_dataset = pd.read_csv(matchesPath);
orange_cap_dataset = pd.read_csv(orange_capPath);
purple_cap_dataset = pd.read_csv(purple_capPath);
```

```python
# Top 10 bowlers who had the most number of dot balls
dotballs = delivery_dataset[(delivery_dataset["runs_of_bat"] == 0)
& (delivery_dataset["extras"] == 0)
& (delivery_dataset["wide"] == 0)
& (delivery_dataset["legbyes"] == 0)
& (delivery_dataset["byes"] == 0)
& (delivery_dataset["noballs"] == 0)];
most_dot_balls = dotballs.groupby("bowler").size().reset_index(name="Number of dot balls").sort_values(by="Number of dot balls" ,ascending=False).head(10).reset_index(drop=True);
most_dot_balls
```

```python
# Most wide and no balls
noballs = delivery_dataset[(delivery_dataset["wide"] != 0) | (delivery_dataset["noballs"] != 0)]
# noballs.shape
most_extras = noballs.groupby("bowler").size().reset_index(name="Number of extras").sort_values(by="Number of extras", ascending=False).head(10).reset_index(drop=True);
most_extras
```

```python
# Most bowling friendly pitch/venue
wicket_deliveries = delivery_dataset[delivery_dataset["wicket_type"].notna()]
wicket_deliveries.shape
```

```python
venuesByWickets = wicket_deliveries.groupby("venue").size().reset_index(name="Wickets fallen").sort_values(by="Wickets fallen", ascending=False).reset_index(drop=True);
venueByMatchesPlayed = delivery_dataset.groupby("venue")["match_no"].nunique().reset_index(name="Number of matches played").sort_values(by="Number of matches played", ascending=False).reset_index(drop=True);
df = pd.merge(venueByMatchesPlayed, venuesByWickets, on="venue");
df
```

```python
# Number of matches won batting first
battingFirstWhenTossLost = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team1']) & (matches_dataset['toss_decision'] == 'Bowl') & (matches_dataset['team2'] == matches_dataset['match_winner'])];
battingFirstWhenTossWon = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team1']) & (matches_dataset['toss_decision'] == 'Bat') & (matches_dataset['team1'] == matches_dataset['match_winner'])];
battingFirstWhenTossLost2 = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team2']) & (matches_dataset['toss_decision'] == 'Bowl') & (matches_dataset['team2'] == matches_dataset['match_winner'])];
battingFirstWhenTossWon2 = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team2']) & (matches_dataset['toss_decision'] == 'Bat') & (matches_dataset['team1'] == matches_dataset['match_winner'])];
# print(battingFirstWhenTossLost.shape)
# print(battingFirstWhenTossWon.shape)
# print(battingFirstWhenTossWon2.shape)
# print(battingFirstWhenTossLost2.shape)
wonBattingFirst = pd.concat([battingFirstWhenTossLost, battingFirstWhenTossWon, battingFirstWhenTossLost2, battingFirstWhenTossWon2]).reset_index(drop=True)
print(wonBattingFirst.shape)
wonBattingFirst.head()
```

```python
# Number of matches won bowling first
bowlingFirstWhenTossWon = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team1']) & (matches_dataset['toss_decision'] == 'Bowl') & (matches_dataset['team1'] == matches_dataset['match_winner'])];
bowlingFirstWhenTossLost = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team1']) & (matches_dataset['toss_decision'] == 'Bat') & (matches_dataset['team2'] == matches_dataset['match_winner'])];
bowlingFirstWhenTossWon2 = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team2']) & (matches_dataset['toss_decision'] == 'Bowl') & (matches_dataset['team1'] == matches_dataset['match_winner'])];
bowlingFirstWhenTossLost2 = matches_dataset[(matches_dataset['toss_winner'] == matches_dataset['team2']) & (matches_dataset['toss_decision'] == 'Bat') & (matches_dataset['team2'] == matches_dataset['match_winner'])];
# print(bowlingFirstWhenTossLost.shape)
# print(bowlingFirstWhenTossWon.shape)
# print(bowlingFirstWhenTossLost2.shape)
# print(bowlingFirstWhenTossWon2.shape)
wonBowlingFirst = pd.concat([bowlingFirstWhenTossLost, bowlingFirstWhenTossWon, bowlingFirstWhenTossLost2, bowlingFirstWhenTossWon2]).reset_index(drop=True)
print(wonBowlingFirst.shape)
wonBowlingFirst.head()
```
