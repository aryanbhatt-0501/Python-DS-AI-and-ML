```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('cricket_players.csv')
```

```python
df.head()
```

```python
df.describe()
```

### Task - 2 Basic Visualisation

```python
runPerCountry = df[['Country', 'Runs']].groupby('Country').sum()
```

```python
runPerCountry
```

```python
sns.barplot(data=runPerCountry, x='Country', y='Runs')
plt.title('Runs per country')
plt.xlabel('Countries')
plt.ylabel('Runs scored')
plt.show()
```

```python
avgPerCountry = df[['Average', 'Country']].groupby('Country').mean()
```

```python
avgPerCountry
```

```python
sns.boxplot(data=df, x='Country', y='Average')
plt.title('Distribution of average per country')
plt.show()
```

### Task - 3 Correlation heatmap

```python
corr_matrix = df.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.tight_layout()
plt.show()
```

```python
sns.barplot(data=df, x='Country', y='Wickets', palette='Set2')
plt.title('Wickets per country')
plt.xlabel('Countries')
plt.ylabel('Wickets taken')
plt.grid(True)
plt.show()
```

```python
sns.scatterplot(data=df, x='Matches Played', y='Runs')
plt.title('Runs vs matches played')
plt.xlabel('Number of matches')
plt.ylabel('Runs scored')
plt.show()
```

### How does seaborn handle categorical and numerical variables differently when plotting ?
#### Categorical = Recognizes x as categorical → groups data by that category.
#### Plots distribution or aggregated value (mean, median, etc.) of y for each category.
#### These plots display variation across categories.
#### Numerocal - Plots relationships or trends between continuous values.
#### Can add regression lines, confidence intervals, etc.

#### From the box plot we can make out that the India has the highest median average
